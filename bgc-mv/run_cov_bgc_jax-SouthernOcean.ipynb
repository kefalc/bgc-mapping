<h1>full SO</h1>

```python
# map Jan and Jun 2017 only - CHANGE THIS IF WANT MORE MONTHS
time_i = (np.datetime64('2017-01-01') - np.datetime64('1970-01-01T00:00:00', 'ns')) / np.timedelta64(1, 'ns') / 1e9 / 86400.0
time_f = (np.datetime64('2017-02-01') - np.datetime64('1970-01-01T00:00:00', 'ns')) / np.timedelta64(1, 'ns') / 1e9 / 86400.0
interval_days = 5
maptimes1 = np.arange( time_i+interval_days/2, time_f, interval_days )

time_i = (np.datetime64('2017-06-01') - np.datetime64('1970-01-01T00:00:00', 'ns')) / np.timedelta64(1, 'ns') / 1e9 / 86400.0
time_f = (np.datetime64('2017-07-01') - np.datetime64('1970-01-01T00:00:00', 'ns')) / np.timedelta64(1, 'ns') / 1e9 / 86400.0
interval_days = 5
maptimes2 = np.arange( time_i+interval_days/2, time_f, interval_days )

maptimes = np.concatenate((maptimes1, maptimes2), axis=0)

maskfile = '/global/homes/k/kefalc/code/gpr-mapping-covar/SO_mask_40S.nc'
ocean_mask = xr.open_dataset(maskfile)
```

<h2>T and O2 500 db</h2>

```python
filename = '/global/homes/k/kefalc/code/gpr-mapping-covar/plevel4_residuals.npz'
outfile = '/global/homes/k/kefalc/code/gpr-mapping-covar/plevel4_T_O2_SouthernOcean.nc'

gridded_tasks_dask(client, outfile, filename, ocean_mask, return_params=True, 
                   correlated_nugget=True, nugget_prime=False, return_estim=True, 
                   winsize_TS=7., winsize_BGC=7., estimtimes=maptimes, verbose=False)

print('Finished with gridded_tasks')
```
